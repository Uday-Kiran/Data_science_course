It is sometimes called Anderson's Iris data set because Edgar Anderson collected the data to quantify the morphologic variation of Iris flowers of three related species.

# Central Limit Theorem

The Central Limit Theorem (CLT) is a statistical theory states that given a sufficiently large sample size from a population with a finite level of variance, the mean of all samples from the same population will be approximately equal to the mean of the population.

```python
import pandas as pd

'''Let's create a dataset called salaries with the lowest being 10000 and the highest being 100000
    and having a row of 300000. And fitting it into a dataset'''
salaries = {'amount' : pd.Series(np.random.randint(10000, 100000, 30000))}
salaries = pd.DataFrame(salaries, index=None)
```

```python
import random
import numpy as np
import seaborn as sb


#initialise a variable incomeData 

incomeData = salaries.amount

#Creating a function to create samples
                        
def getSample(population, sampleSize):
    sampleList = population.sample(sampleSize, replace=True)
    return sampleList

#create a function for mean

def getMean(sampleList):
    sampleSize = len(sampleList)
    totalSumIncome = 0
    for row in sampleList:            
        totalSumIncome = row + totalSumIncome
    mean = totalSumIncome/sampleSize
    return mean

#create a function for Skew and Kurtosis

def getSkewAndKurtosis(sampleList):
    sampleSize = len(sampleList)
    
    totalSumIncome = 0
    for row in sampleList:
        totalSumIncome = row + totalSumIncome
    mean = totalSumIncome/sampleSize
    
    sumOfSquares = 0
    s3 =0
    s4 = 0
    for row in sampleList:
        deviationScore = row - mean
        sumOfSquares = deviationScore**2 + sumOfSquares
        s3 = deviationScore **3 + s3
        s4 = deviationScore **4 + s4
        
    variance = sumOfSquares/(sampleSize - 1)
    
    SD = variance**0.5
    
    n = sampleSize
    s2 = sumOfSquares
    sampleKurtosis = ((n*(n+1))/((n-1)*(n-2)*(n-3)))*((n-1)**2)*(s4/(s2**2))
    
    sampleSkew = s3/((n-1)*SD**3)
    return sampleSkew, sampleKurtosis



x = 0
n = 10 #sample size

statsOfSamples = []
numOfSample = 300

while x < numOfSample:
    x+=1
    mySample = getSample(incomeData, n)
    mySampleMean = getMean(mySample)
    statsOfSamples.append(mySampleMean)
    
    
#calling the functions
popMean = getMean(incomeData)
meanOfMean = getMean(statsOfSamples)
skew, kurtosis = getSkewAndKurtosis(statsOfSamples)
popSkew, popKurtosis = getSkewAndKurtosis(incomeData)

#printing the functions

print('Population Mean', popMean)
print('Mean of Means:', meanOfMean)
print('Skew:',skew)
print('Kurtosis', kurtosis)
print('Population Kurtosis:',popKurtosis)
print('Population Kurtosis:',popSkew)

# plt.hist(statsOfSamples, bins = 50)
sb.distplot(statsOfSamples, bins= 50)
plt.show()


"""keep increasing the the numofSample and you will see that the it starts becoming a normal 
   distribution"""
```
