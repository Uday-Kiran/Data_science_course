```python
x = []
y = []
i = -4
```

```python
while i<4:
    i+=0.01
    x.append(round(i,3))
    
e = 2.718
pi = 3.1415

c = 1/(2*pi)**0.5 

for a in x:
    expo = -(a**2)/2
    
    distro = c*(e**expo)
    y.append(distro)
```

```python
x
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.show()
```

```python
import scipy.integrate

#lower limit
neg_infi = float('inf')

#upper limit
z=0

#function of x
def f(x):
    c = (1/(2*pi)**0.5 )
    exponent = (-x**2)/2
    standard_normal_curve = c*(e**exponent)
    
    return standard_normal_curve


probability, error = scipy.integrate.quad(f, neg_infi, z)

print(round(probability,5))
```

```python
z_table = []

for row in x:
    probability, error = scipy.integrate.quad(f, neg_infi, row)
    z_table.append([row, round(probability, 5)])
    print(row, round(probability, 5))
```

```python
import csv
with open('z_table.csv', 'w', newline='') as f:
    f = csv.writer(f, delimiter = ',')
    for row in z_table:
        f.writerow(row)
```
